```python
#Preditiong rainfall for next 7 days based on date provided
import datetime
import joblib
import numpy as np
import pandas as pd
from statsmodels.tsa.ar_model import AR,ARResults
stations = pd.read_csv('stations.csv')
import warnings
warnings.filterwarnings('ignore')

## Making dataframe with dates to chose index based on date
date_range = pd.date_range(start='1970-01-01',periods=18670,freq='D')
dr = pd.DataFrame(date_range)
year,month,day = input('Enter date in year, month, day format: ').split(',')
ind = dr[dr[0]==datetime.datetime(int(year),int(month),int(day))].index[0]

#if indexis greater than the dataset with which model is fitted then predict from last date fitted in model
if ind>17167:
    start = 17167
else:
    start = ind

##predict and print for all stations

for i in range(1,35):
    vars()['model'+str(i)] = joblib.load('Models\model_'+str(i))
    pred = vars()['model'+str(i)].predict(start=start,end=ind+7)
    pred = abs(pred)
    class color:
        BOLD = '\033[1m'
        UNDERLINE = '\033[4m'
        END = '\033[0m'

    print('                                                                  ')
    print(color.BOLD + color.UNDERLINE + stations.loc[i-1,'0'] + color.END)
    print('                                                                  ')
    print(f'predictions for next seven days for station {stations.loc[i-1,str(0)]} are: ')
    print(pred[-7:])
    print('------------------------------------------------------------------')
```
